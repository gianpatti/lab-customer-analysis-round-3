```python
import pandas as pd
```

```python
data = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')
```

```python
data
```

```python
# 1. Show the dataframe shape.
data.shape
```

```python
# 2. Standardize header names.
data.columns = [col_name.lower().replace(' ', '_') for col_name in data.columns]
```

```python
# 3. Which columns are numerical?
numericals=data.select_dtypes("number")
numericals
```

```python
# 4. Which columns are categorical?
categorical=data.select_dtypes("object")
categorical
```

```python
# 5. Check and deal with `NaN` values. 
data.isna().sum()
```

```python
#Extract the months from the dataset and store in a separate column.
data['month']=pd.DatetimeIndex(data['effective_to_date']).month
```

```python
# Filtering to show only data from first quarter (months 1,2,3)
data=data[data['month']<=3]
data
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
plt.figure(figsize=(12,8))
sns.countplot('sales_channel', hue='response', data=data)
```

```python
plt.figure(figsize=(16,8))
ax1=sns.histplot(data=data, x="total_claim_amount", hue="response", multiple="stack", bins=15, stat='percent')
for i in ax1.containers:
    ax1.bar_label(i,)
```

```python
plt.figure(figsize=(16,8))
ax=sns.histplot(data=data, x="income", hue="response", multiple="stack", bins=15, stat='percent')
for i in ax.containers:
    ax.bar_label(i,)
```
